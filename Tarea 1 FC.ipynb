## 2.- Haga las gráficas de las siguientes funciones de la función $f(x) = tan(x)$ en el intervalo $[−4\pi, 4\pi]$. La gráfica no deben tener líneas verticales.

```python
x = arange(-4*pi,4*pi,0.1) #Creamos un arreglo con los elementos entre -4pi y 4pi con incrementos de 0.1
y=tan(x) #Definimos la función de tangente con los arreglos dados en la linea anterior, esta esta completa y con lineas verticales
for i in range(len(x)):
    if y[i]-y[i-1]<0: #Los puntos que forman las rectas verticales son aquellos que su diferencia es negativa (Prueba con tan(-5)-tan(-4))
        y[i]=nan #Si su diferencia es negativa, osea, generan rectas, los eliminamos con "nan" = not a number, que no son graficados por la función plot
        
plot(x,y,label = "Tan(x)") #Graficamos nuestra nueva función tangente "modificada"
#plot(x,tan(x),"--",label = "Tan(x) completa") #sin alterar
ylim(-3,3)
grid(),legend()
show()
```
