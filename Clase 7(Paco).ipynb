
$$\int_{x_i}^{x_{i+1}}f(x)dx = \frac{h}{48}{23f_{i}+26f_{i+1}-4f_{i+1}+6f_{i+3}-3f_{i+4}}$$

```python
def primitiva(x,y):
    Iy, h = zeros(len(y)),x[1]-x[0]
    for i in range(1,len(y)-3):
        Iy[i] = 23*y[i-1]+26*y[i]-4*y[i+1]+6*y[i+2]-3*y[i+3]+Iy[i-1]
    for i in range(3,0,-1):
        Iy[-i] = 23*y[i]+26*y[-i-1]-4*y[-i-2]+6*y[-i-3]-3*y[-i-4]+Iy[-i-1]
    return Iy*h/48
```

$$
F(x) = \int_0^x \sin(t)dt =-\cos(x)+1
$$

```python
x = linspace(0,2*pi)
y = sin(x)
YY = -cos(x)+1
YY_num = primitiva(x,y)
plot(x,y,label="$(x)$")
plot(x,YY,label="$F(x)$")
plot(x,YY_num,label="$F(x) numèrica$")
legend(),grid()
show()
```
