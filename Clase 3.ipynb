```python
def mi_factorial(n):
    # ojo n es enteroo
    if n == 0:
        fac = 1
    else:
        fac = 1
        for i in range(1,n+1):
            fac =  fac*i   
    return fac

def mi_sin(x):
    ss = 0
    for i in range(30):
        ss += (-1)**(i) * ( x**(2*i+1) / mi_factorial(2*i+1) )
        return ss
```

```python
from pylab import *
from scipy.special import *
#%matplotlib qt4 #es para que las graficas salgan más bonitas
```

```python
def mi_sin(x,N_max=10): #Hay algo raro con la funci{on y eso afecta los resultados
    ss = zeros_like(x)
    for i in range(N_max):
        ss += (-1)**(i) * ( x**(2*i+1) / factorial(2*i+1)) 
        return ss
```

```python
print(sin(10*pi))
print(mi_sin(10*pi))
```

```python
x= linspace(0,3*pi,100) #?linspace para ver qué significan los argumentos
y_np=sin(x)
y_est=mi_sin(x,5)
```

```python
plot(x,y_np,label='Numpy $sin(x)$')
plot(x,y_est,label='Mi $sin(x)$')
ylim([-1.5,1.5])
legend()
show()
```

```python
x= linspace(0,3*pi,100)
y=sin(x)
plot(x,y,label='Numpy $\sin(x)$')
for i in range(2,21,2): #Ponemos 21 para que aparezca hasta el 20 en la leyenda
    y_est=mi_sin(x,i) #La 'i' nos da la precisión. Si la aumentamos va a incrementar el tiempo que tarda en ejecutarse
    plot(x,y_est,label='$\sin(x)$ N=%2.2f'%i) #'%' conecta a i con f, el 2.2 le indica 2 numeros antes y despues del punto, f es flotante, s es string
title('Aproximación del sin(x)')
ylim([-1.5,1.5]),xlim([x[0],x[-1]])
grid()
legend(loc='upper right', bbox_to_anchor=(1.35,1))
show()
```

```python
%%timeit
mi_sin(x,5)
```
