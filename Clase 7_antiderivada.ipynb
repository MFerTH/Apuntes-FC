## Clase 7

$$
 \int_{x_i}^{x_{i+1}} f(x)dx=\frac{h}{48}(23f_i + 26f_{i+1} -4f_{i+2} +6f_{i+3} -3f_{i+4})
$$

```python
from pylab import *
```

```python
def primitiva(x,y):
    
    Iy, h = zeros(len(y)), x[1]-x[0]
    
    for i in range(1,len(y)-3): 
        Iy[i]= 23*y[i-1] +26*y[i] -4*y[i+1] +6*y[1+2] -3*y[1+3] + Iy[i-1]
        
    for i in range(3,0,-1):
        Iy[-i]= 23*y[-i] +26*y[-i-1] -4*y[-i-2] +6*y[-i-3] -3*y[-i-4] + Iy[-i-1]
    return Iy*h/48
```

$$
F(x)= \int_{0}^{x} \sin(t)dt = -\cos(x)+1
$$

```python
x=linspace(0,2*pi)
y=sin(x)
YY=-cos(x)+1
YY_num=primitiva(x,y)

plot(x,y,label="$f(x)$")
plot(x,YY,label="$F(x)$")
plot(x,YY_num,label="$F(x) numérica$")

legend(),grid()
show()
```

```python
def prim_sim(x,y):
    Iy, h = zeros(len(y)), x[1]-x[0]
        
    Iy[1]= 23*y[0] +26*y[1] -4*y[2] +6*y[3] -3*y[4]
    
    for i in range(2,10,2):
        Iy[i]= y[i-2] + 4*y[i-1]+y[i] + Iy[i-2]
        
    for i in range(3,10,2):
         Iy[i]= y[i-2] + 4*y[i-1]+y[i] + Iy[i-2]
    
    return Iy*h/3
```

```python
x=linspace(0,pi,20)
y=sin(x)+1
YY=-cos(x)+x+1
YY_num=primitiva(x,y)
YY_num_sim=prim_sim(x,y)

plot(x,y,label="$f(x)$")
plot(x,YY,label="$F(x)$")
plot(x,YY_num,label="$F(x) numérica$")
plot(x,YY_num_sim,'+',label="$F(x) numérica simétrica$")

legend(),grid()
show()
```
