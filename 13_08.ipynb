```julia
x = collect(-2:0.01:2);
y1 = sin.(x);
y2 = cos.(x);
y3 = x;
y4 = x.^2;
```

```julia
plot(x,y1)
```

```julia
plot(size = [800,500]) #tamaño de la gráfica
plot!(x,y1,color=:blue,line=2,label=L"y_1 = \sin(x)") #el signo ! es para pegar una función encima?
plot!(x,y2,color=:red,line=2,label=L"y_2 = \cos(x)",linestyle=:dash)
plot!(x,y3,color=:violet,line=2,label=L"y_3 = x")
plot!(x,y4,color=:black,line=2,label=L"y_4 = x^2",linestyle=:dash)
title!("Gráfica inicial")
xlabel!("x")
ylabel!("y")
xlims!(-3,3)
ylims!(-3,3)
#savefig("figura-1.png")
```

```julia
x = collect(-3:0.1:3);
y1 = sin.(x);
y2 = cos.(x);
y3 = x;
y4 = x.^2;
```

```julia
anim = @animate for i in 3:length(x)
plot(size = [800,500])
plot!(x[1:i],y1[1:i],color=:blue,line=2,label=L"y_1 = \sin(x)")
plot!(x[1:i],y2[1:i],color=:red,line=2,label=L"y_2 = \cos(x)",linestyle=:dash)
plot!(x[1:i],y3[1:i],color=:violet,line=2,label=L"y_3 = x")
plot!(x[1:i],y4[1:i],color=:black,line=2,label=L"y_4 = x^2",linestyle=:dash)
title!("Animación inicial")
xlabel!("x")
ylabel!("y")
xlims!(-3.5,3.5)
ylims!(-3.5,3.5)
end
gif(anim)
```
